```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df=pd.read_csv("tely10.csv")
df.head()
```

### Data preparation for Linear Regression analysis

```python
df.info()
```

```python
df=df.sort_values(by='year')

seen_cities = set()
new_cities_by_year = {}  
for year in df['year'].unique():
    current_year_cities = set(df[df['year'] == year]['city'])
    new_cities = current_year_cities - seen_cities
    new_cities_by_year[year] = len(new_cities)  
    seen_cities.update(current_year_cities)

# Convert the dictionary to a DataFrame for easier merging
df_new_cities = pd.DataFrame(list(new_cities_by_year.items()), columns=['year', 'new_cities'])

df_new_cities

```

```python
# data is clearn that is a good sign 

#geographical reach per year 
Geog_reach=df.groupby("year")["city"].nunique().reset_index() 
Geog_reach 

Geog_reach['new coming']=df_new_cities['new_cities']
Geog_reach.sort_values(by="year",ascending=False)
```

```python
from sklearn.linear_model import LinearRegression 

x=Geog_reach['year'].values.reshape(-1,1) 
y=Geog_reach['city'].values 

model=LinearRegression() 
model.fit(x,y)#we have now trained the model 


slope=model.coef_[0] 
```

### make predictions for the next two years 

```python
x_predict_2020 = np.array([[2020]])
y_predict_2020 = model.predict(x_predict_2020)
 
another_x=Geog_reach['year'].values.reshape(-1,1) 
another_y=Geog_reach['new coming'].values 

model1=LinearRegression() 
model1.fit(another_x,another_y) 

x_predict_nc = np.array([[2020]])
y_predict_nc = model1.predict(x_predict_2020)

Geog_reach.loc[2020, 'city']=y_predict_2020  # set last point to 2 pi
Geog_reach.loc[1440, 'new coming']=y_predict_nc
```

```python
import matplotlib.image as mpimg


img = mpimg.imread("pic2.jpg")

years = Geog_reach['year']
total_unique_cities = Geog_reach['city']
new_cities = Geog_reach['new coming']

fig, ax = plt.subplots(figsize=(10, 8))

# Adjusting extent to cover the plot area based on your data
left = min(years) - 0.5  # Adding a buffer for visual appeal
right = max(years) + 0.5
bottom = 0  # Assuming you want the image to start from the bottom of the plot
top = max(max(total_unique_cities), max(total_unique_cities - new_cities)) * 1.1  # Adding 10% as a buffer

# Adjusted color scheme for improved clarity and accessibility
total_color = '#3182bd'  # Blue for total unique cities
new_color = '#fd8d3c'  # Orange for new cities

# Setting the background image
ax.imshow(img, aspect='auto', extent=[left, right, bottom, top], zorder=0)

# Plotting bars with adjusted colors
for i, (year, total, new) in enumerate(zip(years, total_unique_cities, new_cities)):
    if year==2020: 
        ax,bar(ax.bar(year, total, color='green', edgecolor='black', width=0.6)
    else:
        ax.bar(year, total, color=total_color, edgecolor='black', width=0.6)
        ax.bar(year, new, bottom=total-new, color=new_color, edgecolor='black', hatch='/', width=0.6)

# Customizing the plot with the new color scheme
ax.set_title('Geographical Reach of the Tely 10 Mile Road Race (2005-2019)', fontsize=18, fontweight='bold')
ax.set_xlabel('Year', fontsize=16)
plt.xticks(years, rotation=45)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Highlighting and annotations
highlight_color = '#d95f02'  # This remains for highlighting significant data points
max_new_cities_year = Geog_reach.loc[Geog_reach['new coming'].idxmax(), 'year']
max_new_cities_value = Geog_reach['new coming'].max()
ax.bar(max_new_cities_year, max_new_cities_value, bottom=total_unique_cities[years==max_new_cities_year]-max_new_cities_value, color=highlight_color, edgecolor='black', hatch='/', label='Most New Cities', width=0.6)
ax.legend()

for i, txt in enumerate(new_cities):
    ax.annotate(txt, (years[i], total_unique_cities[i] + 5), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
